```python
# 讀取所有資料夾下的txt檔
import glob
import os
import pandas as pd
fnames = glob.glob("chinese_news_trans/*/*.[tT][xX][tT]")

datas = {
    "content":[],
    "target":[]    
}

for fn in fnames:
    with open(fn, "r", encoding="utf8") as f:
        content = f.read()
    datas["content"].append(content)
    dn = os.path.split(fn)[0]
    dn = os.path.split(dn)[-1]
    datas["target"].append(dn)
train_df = pd.DataFrame(datas)
train_df
```

```python
fnames = glob.glob("chinese_news_test/*/*.[tT][xX][tT]")

datas = {
    "content":[],
    "target":[]    
}

for fn in fnames:
    with open(fn, "r", encoding="utf8") as f:
        content = f.read()
    datas["content"].append(content)
    dn = os.path.split(fn)[0]
    dn = os.path.split(dn)[-1]
    datas["target"].append(dn)
test_df = pd.DataFrame(datas)
test_df
```

```python
# 答案: 處理成數字
# replace(): 將文字替換成數字
u = train_df["target"].unique()
trans = {target:i for i, target in enumerate(u)}
revers_trans = {i:target for target, i in trans.items()}
print(trans)
print(revers_trans)
y_train = train_df["target"].replace(trans)
y_test = test_df["target"].replace(trans)
y_test
```

```python
# 如果是詩詞分類，標點符號換行，應該去掉，因為它不影響結果。
import jieba
def newscut(s):
    s = " ".join(jieba.cut(s))
    s = s.replace("\r", "").replace("\n", "")
    return s
# 超重要: pandas表格: DataFrame 一欄或一列: Series
# Series.apply(功能名字)
x_train = train_df["content"].apply(newscut)
x_test = test_df["content"].apply(newscut)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
# 訓練資料需要fit+transform，測試資料只需要transform
x_train_vec = vec.fit_transform(x_train)
x_test_vec = vec.transform(x_test)
```

```python
# 單純貝氏
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_vec)
print("正確率:", accuracy_score(pre, y_test))
```

```python
# 混淆矩陣
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pre)
c = ["{}(預測)".format(target) for target in u]
i = ["{}(原本)".format(target) for target in u]
pd.DataFrame(mat, columns = c, index = i)
```

```python
# 回顧
p = input("輸入一篇新聞: ")
# 注意[]: [第一首詩, 第二首詩...]
p = vec.transform([newscut(p)])
# 注意: [第一個答案, 第二個答案...]
ans = clf.predict(p)[0]
print("這文章的分類是: ", revers_trans[ans])
# 標出各個分類的機率
prob = clf.predict_proba(p)[0]
for n, prob in sorted(zip(u, prob), key = lambda x:x[1], reverse=True):
    print(n, "的機率: ", prob *100, "%")
```
