```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding = "utf8")
train_df
```

```python
test_df = pd.read_csv("test.csv", encoding = "utf8")
test_df
```

```python
# 找出資料中的空格
train_df.isna().sum()
```

```python
test_df.isna().sum()
```

```python
# 算出中位數，並drop掉不能用中位數的項目
med = train_df.median().drop(["Pclass"])
train_df = train_df.fillna(med)
train_df.isna().sum()
```

```python
test_df = test_df.fillna(med)
test_df.isna().sum()
```

```python
# 補最常出現
most = train_df["Embarked"].value_counts().idxmax()
train_df["Embarked"] = train_df["Embarked"].fillna(most)
train_df.isna().sum()
```

```python
# 處理name
# # 高手(裝B)寫法
# import re
# s = "Braund, Mr. Owen Harris"
# pat = "(\w+),\s*(\w+)\.(.+)"
# re.search(pat, s).group(3)

s = "Braund, Mr. Owen Harris"
s = s.split(",")[-1].split(".")[0]
# 去除不必要的空格、逗號...等等
s.strip()
```

```python
def namecut(s):
    s = s.split(",")[-1].split(".")[0]
    return s.strip()
# 算出現的數次
train_df["Name"].apply(namecut).value_counts()
```

```python
def namecut(s):
    reserved = ["Mr", "Mrs", "Miss", "Master"]
    s = s.split(",")[-1].split(".")[0]
    s = s.strip()
    if s in reserved:
        return s
    else:
        return None
train_df["Name"] = train_df["Name"].apply(namecut)
train_df
```

```python
test_df["Name"] = test_df["Name"].apply(namecut)
test_df
```

```python
train = pd.get_dummies(train_df, columns = ["Name", "Sex", "Embarked"])
predict = pd.get_dummies(test_df, columns = ["Name", "Sex", "Embarked"])
train
```

```python
x_train = train.drop(["PassengerId", "Survived", "Ticket", "Cabin"], axis = 1)
y_train = train["Survived"]

x_predict = predict.drop(["PassengerId", "Ticket", "Cabin"], axis = 1)
predict_id = predict["PassengerId"]
x_predict
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
clf = RandomForestClassifier()
params = {
    "n_estimators":range(30, 41),
    "max_depth": range(5, 10)
}
grid = GridSearchCV(clf, params, cv = 10, n_jobs=4)
grid.fit(x_train, y_train)
print("最佳參數:", grid.best_params_)
print("最佳分數(10次平均): ", grid.best_score_)
```

```python
from sklearn.model_selection import cross_val_score
clf = RandomForestClassifier(n_estimators=27, max_depth=8)
scores = cross_val_score(clf, x_train, y_train, cv=10, n_jobs=4)
print("十次分數: ", scores)
print("平均分數: ", sum(scores)/len(scores))
```

```python
clf = RandomForestClassifier(n_estimators=27, max_depth=8)
clf.fit(x_train, y_train)
```

```python
pre =  clf.predict(x_predict)
result = pd.DataFrame({
    "PassengerId": predict_id,
    "Survived": pre
})
result.to_csv("titanic_rf.csv", encoding="utf-8", index = False)
result
```

```python
# 5/7 feature_importances_
# zip & key用法很重要，一定要記

imp = clf.feature_importances_
d = sorted(zip(x_train.columns, imp), key = lambda x:x[1], reverse = True)
pd.DataFrame({
    "Name" : [n for n, im in d],
    "Importance" : [im for n, im in d]
})
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_norm = scaler.fit_transform(x_train)
x_predict_norm = scaler.transform(x_predict)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
params = {
    "n_neighbors":range(5, 200)
}
grid = GridSearchCV(clf, params, cv = 10, n_jobs=4)
grid.fit(x_train_norm, y_train)
print("最佳參數:", grid.best_params_)
print("最佳分數(10次平均): ", grid.best_score_)
```

```python
pre = grid.best_estimator_.predict(x_predict_norm)
result = pd.DataFrame({
    "PassengerId": predict_id,
    "Survived": pre
})
result.to_csv("titanic_knn.csv", encoding="utf-8", index = False)
result
```
